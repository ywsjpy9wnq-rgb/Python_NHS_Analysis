```python
import pandas as pd

waiting = pd.read_csv('/Users/matthewgibbs/Documents/Work/Current/Portfolio/Coding/nhs_database/data/nhs_waiting_lists.csv')
staffing = pd.read_csv('/Users/matthewgibbs/Documents/Work/Current/Portfolio/Coding/nhs_database/data/nhs_staffing_info.csv')

#Merge
merged = pd.merge(waiting, staffing, on='organisation_code', how='inner')

#Define bins and labels
patient_bins = [0, 10000, 25000, 50000, float('inf')]
patient_labels = ['0-10k', '10k-25k', '25k-50k', '>50k']
doctor_bins = [0, 500, 1000, 2000, float('inf')]
doctor_labels = ['0-500', '501-1000', '1001-2000', '>2000']

#Create categorical columns
merged['patient_group'] = pd.cut(merged['up_to_18_weeks'], bins=patient_bins, labels=patient_labels)
merged['doctor_group'] = pd.cut(merged['doctors'], bins=doctor_bins, labels=doctor_labels)

#Crosstab
table = pd.crosstab(merged['patient_group'], merged['doctor_group'], margins=True)
print(table)
```
