```python
import pandas as pd
import statsmodels.api as sm

waiting = pd.read_csv('/Users/matthewgibbs/Documents/Work/Current/Portfolio/Coding/nhs_database/data/nhs_waiting_lists.csv')
staffing = pd.read_csv('/Users/matthewgibbs/Documents/Work/Current/Portfolio/Coding/nhs_database/data/nhs_staffing_info.csv')

# Merge
merged = pd.merge(waiting, staffing, on='organisation_code', how='inner')

# Create binary outcome: 1 if any patient waited >52 weeks
merged['long_wait'] = (merged['over_52_and_up_to_65_weeks'] > 0).astype(int)

# Select predictors (staff groups)
predictors = ['doctors', 'nurses_and_health_visitors', 'midwives', 
              'ambulance_staff', 'managers', 'others']

X = merged[predictors].fillna(0)   
X = sm.add_constant(X)
y = merged['long_wait']

# Fit logistic regression
model = sm.Logit(y, X).fit()
print(model.summary())
```
