```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/Users/matthewgibbs/Documents/Work/Current/Portfolio/Coding/nhs_database/data/nhs_waiting_lists.csv')

plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='total_waiting_list', y='up_to_18_weeks', alpha=0.6)
plt.xlabel('Total waiting list size')
plt.ylabel('Number of patients waiting ≤18 weeks')
plt.title('Relationship between total waiting list and timely care')
plt.grid(True, linestyle='--', alpha=0.3)
plt.show()

#Interpretation
print("There is a strong positive correlation: organisations with larger total waiting lists also tend to have more patients waiting within the 18-week target. No clear outliers.")
```
