```python
import pandas as pd

df = pd.read_csv('/Users/matthewgibbs/Documents/Work/Current/Portfolio/Coding/nhs_database/data/nhs_staffing_info.csv')

#1. Number with ambulance staff
has_ambulance = df['ambulance_staff'] > 0
count_amb = has_ambulance.sum()
print(f"Organisations with ambulance staff: {count_amb}")

#2. Average doctors by group
avg_doctors_with = df[has_ambulance]['doctors'].mean()
avg_doctors_without = df[~has_ambulance]['doctors'].mean()

print(f"Average doctors (with ambulance staff): {avg_doctors_with:.1f}")
print(f"Average doctors (without ambulance staff): {avg_doctors_without:.1f}")
```
